```python
import numpy as np
import pandas as pd
```

```python
# loading the data from the csv file to apandas dataframe
df = pd.read_csv('data.csv')
df.head()
```

```python
df.shape
```

```python
df = df[df['year'] >= 2000]  #taking song only after 2000
```

```python
df = df[df['popularity'] >= 65] #taking song with popularity score over 65
```

```python
# resetting the index from 1 t0 ...
df = df.reset_index(drop=True)
df
```

```python
df.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ['acousticness','artists','danceability','energy','instrumentalness','liveness','loudness','mode','tempo','release_date']
selected_features
```

```python
# replacing the null valuess with null string

for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
# combining all the selected features

combined_features = (
    df['acousticness'].astype(str) + ' ' +
    df['artists'] + ' ' +
    df['danceability'].astype(str) + ' ' +
    df['energy'].astype(str) + ' ' +
    df['instrumentalness'].astype(str) + ' ' +
    df['liveness'].astype(str) + ' ' +
    df['loudness'].astype(str) + ' ' +
    df['mode'].astype(str) + ' ' +
    df['popularity'].astype(str) + ' ' +
    df['tempo'].astype(str) + ' ' +
    df['release_date']
)
```

```python
print(combined_features)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# converting the data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
similarity.shape
```

```python
df[df['name']=='Yellow']
```

```python
def recommendation(song_df):
    idx = df[df['name'] == song_df].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])
    
    songs = []
    for m_id in distances[1:6]:
        songs.append(df.iloc[m_id[0]]['name'])
        
    return songs
```

```python
recommendation('Trouble')
```

```python
import pickle
pickle.dump(similarity,open('similarity.pkl','wb'))
```
